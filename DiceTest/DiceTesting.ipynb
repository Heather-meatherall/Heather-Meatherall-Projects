# Dice Testing

Many TTRPGs, such as Dungeons and Dragons, using a D20 systems, which involves rolling a d20 to see if the player is successful at a specific task. 

As a player, you want to maximise how many twenties you roll and minimise the number of ones you roll. There are many different methods to rolling dice, either using physical dice or one of the many online dice generators, and players have spefic rituals they use in an attempt to acheive these results. 

In this project, we are going to compare the different methods in an attempt to see which one is more successful. For the purposes of this project, we will define success as highest number of twenties rolled and lowest number of ones rolled

# Methods
## Data Collecting

Data was gathered by rolled both a physical di and virtual dice.

For physical dice, a plastic d20 was used. It was rolled after sitting with the 20, 1, and 10's side facing up for 30 minutes (referred to as charging). It was also rolled without charging.

For virtual dice, online dice rollers including Dnd Beyond, Discord Dice Roller bot, and Google's dice roller were used.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load data
dice = pd.read_csv('DiceRolls.csv')
```

# Analysis
I will explore the relationships between the different type of dice rolled and the frequency of natural 20s and natural 1s. I will do this with exploratory analysis consisting of statistical summaries and visualizations.

## Averages
I will begin by comparing the average twenties and ones rolled for Physical and Virtual Dice. We can see that on average, twenties and ones were rolled more often by the physical dice. However, we can also see that different between them is not a large interval. 

```python
physicalTwentyAverage = round(dice.loc[dice.Type == 'Physical'].Twenty.mean(), 2)
virtualTwentyAverage = round(dice.loc[dice.Type == 'Digital'].Twenty.mean(), 2)

physicalOneAverage = round(dice.loc[dice.Type == 'Physical'].One.mean(), 2)
virtualOneAverage = round(dice.loc[dice.Type == 'Digital'].One.mean(), 2)

print(f'Physical Dice Twenty Average: {physicalTwentyAverage}')
print(f'Virtual Dice Twenty Average : {virtualTwentyAverage}')
print(f'Physical Dice One Average: {physicalOneAverage}')
print(f'Virtual Dice One Average : {virtualOneAverage}')
```

Next I looked at different ways the Physical dice was rolled.
We can see that the di charged with the 10 facing up rolled more twenties and less ones then the others

```python
unchargedTwenty = round(dice.loc[dice.Subtype == 'uncharged'].Twenty.mean(), 2)
highestTwenty = round(dice.loc[dice.Subtype == 'highest number'].Twenty.mean(), 2)
lowestTwenty = round(dice.loc[dice.Subtype == 'lowest number'].Twenty.mean(), 2)
averageTwenty = round(dice.loc[dice.Subtype == 'average'].Twenty.mean(), 2)

unchargedOne = round(dice.loc[dice.Subtype == 'uncharged'].One.mean(), 2)
highestOne = round(dice.loc[dice.Subtype == 'highest number'].One.mean(), 2)
lowestOne = round(dice.loc[dice.Subtype == 'lowest number'].One.mean(), 2)
averageOne = round(dice.loc[dice.Subtype == 'average'].One.mean(), 2)

print("Nat Twenties")
print(f'Uncharged: {unchargedTwenty}')
print(f'Charged with 20 facing up: {highestTwenty}')
print(f'Charged with 1 facing up: {lowestTwenty}')
print(f'Charged with 10 facing up: {averageTwenty}\n')

print("Nat Ones")
print(f'Uncharged: {unchargedOne}')
print(f'Charged with 20 facing up: {highestOne}')
print(f'Charged with 1 facing up: {lowestOne}')
print(f'Charged with 10 facing up: {averageOne}')
```

I then looked at different virtual dice rollers that were used. We can see that the Dnd Beyond dice bot rolls more twenties and ones then the others

```python
discordTwenty = round(dice.loc[dice.Subtype == 'discord bot'].Twenty.mean(), 2)
dndTwenty = round(dice.loc[dice.Subtype == 'dnd beyond'].Twenty.mean(), 2)
googleTwenty = round(dice.loc[dice.Subtype == 'google'].Twenty.mean(), 2)

discordOne = round(dice.loc[dice.Subtype == 'discord bot'].One.mean(), 2)
dndOne = round(dice.loc[dice.Subtype == 'dnd beyond'].One.mean(), 2)
googleOne = round(dice.loc[dice.Subtype == 'google'].One.mean(), 2)

print("Nat Twenties")
print(f'Discord Bot: {discordTwenty}')
print(f'Dnd Beyond: {dndTwenty}')
print(f'Google Dice Roller: {googleTwenty}\n')

print("Nat ones")
print(f'Discord Bot: {discordOne}')
print(f'Dnd Beyond: {dndOne}')
print(f'Google Dice Roller: {googleOne}')
```

Finally, I compared the di charged with the 10 facing up and the Dnd Beyond dice roller. As we can see, their results were identical

```python
print('Twenties')
print(f'Charged with 10 facing up: {averageTwenty}')
print(f'Dnd Beyond: {dndTwenty}\n')

print('Ones')
print(f'Charged with 10 facing up: {averageOne}')
print(f'Dnd Beyond: {dndOne}')
```

The figures below also demonstate this

```python
physicalAverage = round(dice.loc[dice.Type == 'Physical'].mean(), 2)
virtualAverage = round(dice.loc[dice.Type == 'Digital'].mean(), 2)

x = np.arange(1,21)
data1 = physicalAverage[0:20]
data2 = virtualAverage[0:20]

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Dice Roll Averages')
ax1.bar(x, data1)
ax2.bar(x, data2)
ax1.set_title("Physical Dice Averages")
ax1.set_xlabel("Number Rolled")
ax1.set_ylabel("Number of Times Rolled (Average)")

ax2.set_title("Virtual Dice Averages")
ax2.set_xlabel("Number Rolled")
ax2.set_ylabel("Number of Times Rolled (Average)")

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.show()


```

```python
unchargedAverage = round(dice.loc[dice.Subtype == 'uncharged'].mean(), 2)
highestAverage = round(dice.loc[dice.Subtype == 'highest number'].mean(), 2)
lowestAverage = round(dice.loc[dice.Subtype == 'lowest number'].mean(), 2)
tenAverage = round(dice.loc[dice.Subtype == 'average'].mean(), 2)


x = np.arange(1,21)
data1 = unchargedAverage[0:20]
data2 = highestAverage[0:20]
data3 = lowestAverage[0:20]
data4 = tenAverage[0:20]

fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4)
fig.suptitle('Physical Dice Roll Averages')
ax1.bar(x, data1)
ax2.bar(x, data2)
ax3.bar(x, data3)
ax4.bar(x, data4)

ax1.set_title("Uncharged Dice Averages")
ax1.set_xlabel("Number Rolled")
ax1.set_ylabel("Number of Times Rolled (Average)")

ax2.set_title("Charged with 20 Dice Averages")
ax2.set_xlabel("Number Rolled")
ax2.set_ylabel("Number of Times Rolled (Average)")

ax3.set_title("Charged with 1 Dice Averages")
ax3.set_xlabel("Number Rolled")
ax3.set_ylabel("Number of Times Rolled (Average)")

ax4.set_title("Charged with 10 Dice Averages")
ax4.set_xlabel("Number Rolled")
ax4.set_ylabel("Number of Times Rolled (Average)")

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=2.0,
                    top=0.9,
                    wspace=0.9,
                    hspace=0.4)


```

# Discussion

In this project I explored data gathered from different dice rolling methods to see if they had an effect on how successful the rolls where. I found that while it doesn't really matter if a physical or virtual di is used, but the spefic type does appear to. I found that on average, the di charged with the 10 facing up and the dnd beyond dice rolled better than the other options.

These outcomes suggest that when a ttrpg plyer is looking to maximuse the number of twenties they roll and minimize the number of ones they roll, they should either roll a physical die charged with the 10 facing up or they should use the dice roller provided by Dnd Beyond.
